```python
import cv2
from binarization import otsu
import matplotlib.pyplot as plt
```

```python
def plt_display(image, title):
    fig = plt.figure()
    a = fig.add_subplot(1, 1, 1)
    imgplot = plt.imshow(image)
    a.set_title(title)
```

# OTSU

```python
img = cv2.imread('examples/0.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
thresh = otsu.threshold_otsu(img, otsu.OtsuType.classic)
binary = img > thresh

plt_display(img, 'original')
plt_display(binary, 'otsu')
print(thresh)
```

```python
thresh = otsu.threshold_otsu(img, otsu.OtsuType.kurita_otsu_abdelmalek)
binary = img > thresh

plt_display(img, 'original')
plt_display(binary, 'otsu')
print(thresh)
```

```python
thresh = otsu.threshold_otsu(img, otsu.OtsuType.kittler_lllingworth)
binary = img > thresh

plt_display(img, 'original')
plt_display(binary, 'otsu')
print(thresh)
```

# Niblack

```python
from binarization import niblack

thresh = niblack.threshold_niblack(img, window_size=55, k=0.1)
binary = img > thresh
plt_display(img, 'original')
plt_display(binary, 'niblack')
```

```python
thresh = niblack.threshold_niblack_const(img, window_size=55, k=0.01, c=30.22)
binary = img > thresh
plt_display(img, 'original')
plt_display(binary, 'niblack')
```
